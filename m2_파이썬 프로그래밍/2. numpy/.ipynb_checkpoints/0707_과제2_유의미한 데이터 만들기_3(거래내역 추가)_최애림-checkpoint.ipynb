## 1. 랜덤 데이터 생성

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime 
import time
```

### 고객 속성 데이터프레임 만들기

```python
random.seed(1)

#고객별 속성 데이터프레임 만들기

cusno=np.arange(1001,2001)
#cusno=random.sample(list(range(1000,9000)),1000)
gender=np.random.choice([0,1],size=1000)
age=np.random.randint(10,81,size=1000)
region=np.random.randint(1,11,size=1000)
```

```python
cusno=pd.Series(cusno)
gender=pd.Series(gender)
age=pd.Series(age)
region=pd.Series(region)

#시리즈 합쳐서 데이터프레임 만들기
customer=pd.concat([cusno,gender,age,region],axis=1)

#컬럼명 지정
customer.columns=['cusno','gender','age','region']

customer.head()
```

```python
#데이터프레임 채우기

purchase['id']=id
purchase['cusno']=cusno
purchase['product']=product
purchase['price']=price
purchase['quantity']=quantity
purchase['amount']=amount


purchase.head()
```

```python
#날짜거래 데이터프레임 만들기

#1.date 
date_list=list(pd.date_range('2020-01-01','2021-12-31',freq='D')) #랜덤 선택을 위한 date 리스트

purchase['date']=np.random.choice(date_list,size=10000)

#2. day(요일) 
purchase['day']=purchase['date'].apply(lambda x: x.weekday())  #date별 요일을 숫자로 나타냄
purchase['day']=purchase['day'].replace(list(range(7)),['월','화','수','목','금','토','일'])

#3. 시간
purchase['time']=np.random.randint(1,25,size=10000)


purchase.head()
```

## 2. 데이터 전처리

```python
#gender value 문자화

df['gender'].replace({0:'male',1:'female'}, inplace=True)
df.head()
```

```python
#region value 문자화

df['region'].replace({1:'서울', 
                      2:'경기',
                      3:'부산',
                      4:'강원',
                      5:'충북',
                      6:'충남',
                      7:'전남',
                      8:'전북',
                      9:'경남',
                      10:'경북',                      
                     },inplace=True)
df.head()
```

```python
#age 범주화

def age_category(x):
    y=0
    if x<20:
        y='10대'
    elif x<30:
        y='20대'
    elif x<40:
        y='30대'
    elif x<50:
        y='40대'
    elif x<60:
        y='50대'
    else:
        y='60대'
    return y

df['age_c']=df['age'].apply(age_category)
df.head()
```

```python
y= round(np.random.uniform(1000,20001),2)
```

```python
#제품코드에 따라 price, quantity, amount, sales 수정

#price, quantity 수정
def price_set(x):
    if x==10: 
        y= np.random.uniform(100000,200001)
    elif x==20:
        y= np.random.uniform(20000,100001)
    elif x==30: 
        y= np.random.uniform(1000,20001)
    return y

def quantity_set(x):
    if x==10: 
        z= np.random.randint(1,21)
    elif x==20:
        z= np.random.randint(1,50)
    elif x==30: 
        z= np.random.randint(1,100)
    return z

df['price']=df['product'].apply(price_set).round(2)
df['quantity']=df['product'].apply(quantity_set)


#amount 수정

df['amount']=df['price']*df['quantity']


df.head()
```

```python
#데이터 순서를 날짜, 시간으로 정렬

df=df.sort_values(['date','time'],ignore_index=True)  #인덱스 새로 생성
df.head()
```

```python
# id 변수를 인덱스화

df.set_index('id',inplace=True)
df.head()
```

### 데이터확인

```python
df['year']=df['date'].dt.year
df.head()
```

```python
#고객별, 연도별 그룹화하여 데이터 탐색
table=pd.pivot_table(df,
                     index='cusno',
                     columns='year',
                     values='amount',
                     aggfunc=['count','sum']
                    )

table.head()
```

```python
#null값 확인: (20년도 구매 o,21년도 구매 x),(20년도 구매 x, 21년도 구매 o)

table.isnull().sum()
```

```python
#null값을 0으로 변경

table.fillna(0,inplace=True)
```

```python
#null값 다시 확인

table.isnull().sum()
```

```python
#변수 데이터타입을 정수로 변경

table['count']=table['count'].astype(int)
```

```python
#총 구매횟수 차이 구하기
table['count_df']=table.iloc[:,1]-table.iloc[:,0]

#총 구매액 차이 구하기
table['amount_df']=table.iloc[:,3]-table.iloc[:,2]

#총 구매액이 증가했는지 여부 확인하는 sales 변수 생성
table['sales']=np.where(table['amount_df']>=0,1,0)

table.head()
```

```python
table.columns=['cusno','count_2020','count_2020','amount_2020','amount_2021',
              'amount_df','sales']
```
