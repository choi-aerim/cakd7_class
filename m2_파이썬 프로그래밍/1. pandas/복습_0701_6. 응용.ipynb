#### Q2. 10을 더하는 함수 add_10 과 두 객체의 합 add_two_obj 함수를 정의한 후 결과값이 각각 20과 20이 되도록 출력하세요.


```python
def add_10(x):
    return x+10

print(add_10(10))
```

```python
def add_two_obj(x,y):
    return x+y

print(add_two_obj(8,12))
```

#### Q3. 10을 곱하는 사용자 함수와 두 객체의 나눗셈을 수행하는 사용자 함수를 정의한 후 결과값이 각각 200과 10을 출력하세요.


```python
def mul_10(x):
    return x*10

print(mul_10(20))
```

```python
def div_two_obj(x,y):
    return x/y

print(div_two_obj(200,20))
```

#### Q8. df_map의 [0,0]을 NaN으로 변경한 후 missing_value 함수를 적용해서 불린 시리즈를 반환하세요.
- 과제 8번에 df_map은 7번 문제에서 df_map으로 이름을 사용하신 후 진행하시면 됩니다.


```python
import numpy as np

df_map.iloc[0,0]=np.nan

df_map.head()
```

```python
#방법1
def m_value(series):
    return series.isnull()

df_map.apply(m_value)
```

```python
#방법2
def missing_value(x):
    try:
        if int(x)>0:
            return False
    except: 
        return True
    
df_map.applymap(lambda x: missing_value(x))
```

#### Q9. titanic 데이터셋을 seaborn에서 로딩한 후 블린 인덱싱을 이용하여 아래 사항들을 수행하세요.
- 나이가 10대(10~19세)인 승객만 따로 선택
- 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
- 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택


```python
titanic.head()
```

```python
titanic.info()
```

```python
#나이가 10대인 승객 선택

tit_10=titanic[(titanic['age']>=10)&(titanic['age']<20)]
tit_10.head()
```

```python
#확인작업
tit_10['age'].min(), tit_10['age'].max()
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
tit_under10_f=titanic[(titanic['age']<10)&(titanic['sex']=='female')]
tit_under10_f.head()
```

```python
#확인작업
age=tit_under10_f['age']
print(age.min(), age.max())

sex=tit_under10_f['sex']
print(set(sex.values))
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택
tit_complex=titanic.loc[(titanic['age']<10)|(titanic['age']>=60),['age','sex','alone']]
tit_complex.head()
```

```python
#확인 작업
set(tit_complex['age'].values)
```
