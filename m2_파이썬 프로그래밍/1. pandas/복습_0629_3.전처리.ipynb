#### -연령 범주화하기

```python
#범주화 전 연령 변수의 value 확인
df.age.value_counts()   
```

```python
# 연령 범주화 함수 작성
def age_cat(x):
    ac=''                          # ac=''가 꼭 들어가야 ac 값을 반환이 가능해진다!!!! 
    if x<20:
        ac='child'
    elif x<30:
        ac='young adult'
    elif x<60:
        ac='adult'
    elif x>=60:
        ac='elderly'
    else:
        pass
    return ac                                       
```

```python
#함수 적용해 새로운 범주 함수 만들기
df.head()
df['age_category']=df.age.map(lambda x: age_cat(x))
df[['age','age_category']].head(10)


df['age_category2']=df.age.apply(lambda x: age_cat(x))
df[['age','age_category2']].head(10)
```

### mtcars 데이터셋의 qsec컬럼을 최소 최대 척도로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하기
- 컬럼들마다 데이터의 크기가 다르기 때문에 편중이 발생할 수 있음
- 따라서, 최대,최소 값을 가지고 정규화를 실행, 객관적 지표로 변환 후, 0.5보다 큰 값을 가지는 레코드 수 구하기

```python
import os
os.getcwd()
```

```python
#데이터셋 불러오기
import pandas as pd

df_auto=pd.read_csv('C:\cakd7\m2_분석라이브러리활용\pandas\dataset\mtcars.csv')
df_auto.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

#모델링에 적합하게 qsec변수 전처리: (-1,1) 배열 형태
###(-1,1)배열은 열을 1개로 해서 자동적으로 2차원 배열을 만들라는 뜻
qsec=df_auto['qsec'].values.reshape(-1,1)
qsec,type(qsec)
```

```python
#MinMaxScaler 정규화: (x-x.min)/(x.max-x.min)
scaler=MinMaxScaler()
scaled=scaler.fit_transform(qsec)
```

```python
#데이터 프레임으로 변환
d=pd.DataFrame(scaled,columns=['qsec_scaled'])
d.head(10)
```

```python
#0.5보다 큰 값을 가지는 레코드 수 구하기
d[d['qsec_scaled']>0.5].count()
```
