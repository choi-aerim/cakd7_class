````python
import numpy as np

# 원본 행렬 R 생성, 분해 행렬 P와 Q 초기화, 잠재요인 차원 k는 3으로 설정

R = np.array([[4, np.nan, np.nan, 2, np.nan],
              [np.nan, 5, np.nan, 3, 1],
              [np.nan, np.nan, 3, 4, 4],
              [5, 2, 1, 2, np.nan]])

# 사용자수, 아이템 수 객체 생성
num_users, num_items = R.shape

# 차원 수
k = 3


# P와 Q 행렬 크기 지정하고 정규분포를 가진 임의의 값으로 입력함
P = np.random.normal(scale = 1./k, size = (num_users, k))
Q = np.random.normal(scale = 1./k, size = (num_items, k))
````

````python
P, Q
````

````python
# 실제 R행렬과 예측행렬의 오차를 구하는 get_rmse 사용자함수 생성

from sklearn.metrics import mean_squared_error

def get_rmse(R, P, Q, non_zeros):
    error = 0
    
    # 두 개의 분해된 행렬 P와 Q.T의 내적으로 예측 R 행렬 생성
    full_pred_matrix = np.dot(P, Q.T)
    
    # 실제 R행렬에서 null이 아닌 값의 위치 인덱스 추출해서 실제 R행렬과 예측 행렬의 RMSE 추출
    x_non_zero_ind = [non_zero[0] for non_zero in non_zeros]
    y_non_zero_ind = [non_zero[1] for non_zero in non_zeros]
    R_non_zeros = R[x_non_zero_ind, y_non_zero_ind]
    
    full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind]
    
    mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
    rmse = np.sqrt(mse)
    
    return rmse
````

````python
# SGD(확률적 경사하강법)기반 행렬분해 수행
## SGD: 과적합을 방지하기 위해 L2규제를 반영해 실제 R행렬 값과 예측 R행렬 값의 차이를 최소화하는 방향성을 가지고(= 경사하강법)
## 행렬과 Q행렬에 업데이트 값을 반복적으로 수행하면서 최적화된 예측 R 행렬을 구하는 방식


# R > 0인 행위치, 열 위치, 값을 non_zeros 리스트에 저장
non_zeros = [(i, j, R[i,j]) for i in range(num_users) for j in range(num_items) if R[i, j] > 0]

steps = 1000
learning_rate = 0.01
r_lambda = 0.01

# SGD 기법으로 P와 Q 매트릭스 계속 업데이트
for step in range(steps):
    for i, j, r in non_zeros:
        # 실제 값과 예측값의 차이인 오류값 구함
        eij = r - np.dot(P[i,:], Q[j,:].T)
        
        # Regularization을 반영한 SGD 업데이트 공식 적용
        P[i,:] = P[i, :] + learning_rate * (eij * Q[j,:].T - r_lambda * P[i,:])
        Q[i,:] = Q[j, :] + learning_rate * (eij * P[i,:].T - r_lambda * Q[i,:])
        
    rmse = get_rmse(R, P, Q, non_zeros)

    if (step % 50) == 0:
        print('***** iteration step: ', step, 'rmse: ',rmse)
````

````python
# 원본행렬 출력
print('원본 행렬: \n', R)

# 분해된 P, Q 함수를 P*Q.T로 예측 행렬을 만들어서 출력

pred_matrix = np.dot(P, Q.T)
print('예측 행렬: \n', np.round(pred_matrix, 3))
````

- 원본 행렬과 비교해 null이 아닌 값은 큰 차이가 나지 않으며, null인 값은 새로운 예측값으로 채워짐

### 장르 콘텐츠 유사도 측정
- 장르 컬럼은 여러개의 개별 장르가 리스트로 구성되어 있음
- 만약 영화1의 장르가 [액션, 모험, 판타지, SF]로 되어있고, 영화2의 장르가 [모험, 판타지, 액션]으로 되어 있다면, 장르별 유사도를 측정할 방법은
    1. 장르를 문자열로 변경 후 <span style = 'background-color:  #fff5b1'>CounterVectorizer로 피처벡터화</span>
    2. 피처벡터화한 행렬 데이터 값을 코사인 유사도로 비교
    3. 장르 유사도가 높은 영화 중 평점이 높은 순으로 영화 추천

#### 1. 장르칼럼을 문자열로 변환

````python
from sklearn.feature_extraction.text import CountVectorizer

# 리스트 -> 공백으로 join한 문자화
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x: (' ').join(x))

# 카운터벡터화
## 카운터벡터 클래스 객체화
count_vect = CountVectorizer(min_df = 0, ngram_range = (1,2))

## 변형
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape, '\n', genre_mat)
````

- 4803개의 레코드와 276개의 개별 단어 피처로 구성된 벡터 행렬이 만들어짐

#### 코사인 유사도 측정

````python
from sklearn.metrics.pairwise import cosine_similarity

# 피처벡터화한 genre_mat 행렬에 대해 코사인 유사도 측정
genre_sim = cosine_similarity(genre_mat, genre_mat)


print(genre_sim.shape)
print(genre_sim[:5])
````

- genre_sim 객체는 데이터의 행별 유사도 정보를 가지고 있음(4803 * 4803이니까)
- 장르 기준으로 콘텐츠 기반 필터링을 수행하려면 movie_df의 개별 레코드에 대해 가장 장르 유사도가 높은 순으로 레코드를 추출해야함

````python
# 행의 유사도가 높은 순으로 정렬된 행렬 위치 인덱스 값 추출하기

genre_sim_ind = genre_sim.argsort()[:,::-1]

print(genre_sim_ind[:5])
````

````python
# 장르 유사도에 따라 영화를 추천하는 함수 생성

def find_sim_movie(df, sorted_ind, title_name, top_n = 10):
    # 인자로 입력된 movie_df에서 title 컬럼이 입력된 title_name 값인 데이터프레임 추출
    title_movie = df[df['title'] == title_name]
    
    # title_name를 가진 데이터프레임의 인덱스 객체를 ndarray로 반환하고
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n개의 인덱스 추출
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :top_n]
    
    # 추출된 top_n 인덱스 출력-> 2차원 array임
    # 데이터 프레임에서 인덱스로 사용하기 위해 1차원 array로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    print(similar_indexes)
    
    return df.iloc[similar_indexes]
````

### 가중치가 적용된 평점 컬럼 생성

```
가중 평점 = (v/(v+m)) * R + (m/(v+m)) * c

    v : 개별 영화에 평점을 투표한 횟수 -> voted_count
    m : 평점을 부여하기 위한 최소 투표 횟수 
        -> m값을 높이면 평점 투표 횟수가 많은 영화에 더 많은 가중 평점을 부여함
    R : 개별 영화에 대한 평균 평점 -> vote_average
    c : 전체 영화에 대한 평균 평점 -> movies_df['vote_average'].mean()
```

````python
# 전체 영화에 대한 평균 평점
C = movies_df['vote_average'].mean()

# 상위 60% 값
m = movies_df['vote_count'].quantile(0.6)
# m = np.percentile(movies_df['vote_count'],60)
C, m
````

````python
# 가중치 계산해주는 사용자함수 생성


C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6)

def weighted_vote_average(x):
    v = x['vote_count']
    R = x['vote_average']
    
    return ((v/(v + m)) * R + (m/(v + m)) * C)


# 원본데이터프레임에 가중치를 적용한 평점 컬럼 생성
movies_df['weight_vote'] = movies_df.apply(weighted_vote_average, axis = 1)
````

````python
# 오름차순 정렬하여 상위 10개 출력

movies_df[['title','vote_average', 'vote_count', 'weight_vote']].sort_values('weight_vote', ascending = False)[:10]
````

#### 새롭게 정의된 평점 기준에 따라서 영화 추천하기
- top_n을 두배수만큼 후보군으로 선정한 뒤에 weight_vote 컬럼값이 높은 순으로 top_n만큼 추출하는 사용자함수 생성

````python
def find_sim_movie(df, sorted_ind, title_name, top_n = 10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배에 해당하는 장르 유사성이 높은 인덱스 추출
    similar_indexes = sorted_ind[title_index, :top_n * 2]
    ## 인덱스의 2차원 데이터를 1차원 데이터로 변형
    similar_indexes = similar_indexes.reshape(-1)
    
    # 기준 영화 인덱스 제외
    similar_indexes = similar_indexes[similar_indexes != title_index]
    
    # top_n의 2배에 해당하는 후보군에서 weighted_vote가 높은 순으로 top_n만큼 추출
    return df.loc[similar_indexes].sort_values('weight_vote', ascending = False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_ind, 'The Godfather', 10)
similar_movies
````

### 데이터불러오기

````python
import pandas as pd
import numpy as np

movies = pd.read_csv('dataset_ml_latest_small/movies.csv')
ratings = pd.read_csv('dataset_ml_latest_small/ratings.csv')

display(movies.head())
display(ratings.head())
````

### 데이터프레임 merge

````python
ratings_movie = pd.merge(movies, ratings, how = 'inner', on = 'movieId')
ratings_movie.head()
````

### 사용자-아이템 간 평점 희소행렬 만들기

#### pivot_table생성

````python
ratings_matrix = pd.pivot_table(data = ratings_movie,
                                index = 'userId',
                                columns = 'title',
                                values = 'rating')
ratings_matrix
````

#### null값을 0으로 채우기

````python
ratings_matrix.fillna(0, inplace = True)

ratings_matrix
````

### 영화 간 유사도 산출

````python
# 아이템(영화)기반 추천하기 위해 데이터프레임 전치
ratings_matrix_T = ratings_matrix.T
ratings_matrix_T.head()
````

````python
from sklearn.metrics.pairwise import cosine_similarity

# 코사인 유사도 산출
item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)
item_sim_df = pd.DataFrame(item_sim, index = ratings_matrix_T.index, columns = ratings_matrix_T.index)

print(item_sim_df.shape)
item_sim_df
````

````python
# 영화 갓파더와 코사인유사도가 높은 10개 영화 추출

item_sim_df.loc['Godfather, The (1972)'].sort_values(ascending = False)[:10]
````

- 갓파더를 제외하면 갓파더2, 좋은친구들, 뻐꾸기 둥지 위로 날아간 새, 스타워즈 등이 유사도가 높게 나타남

    - <span style = 'background-color: #fff5b1'>아직 관람하지 않은 영화에 대해서 <b>아이템 유사도</b>와 <b>기존에 관람한 영화의 평점 데이터</b>를 기반으로 새롭게 모든 영화의 예측 평점을 계산 후, 예측 평점이 높은 영화를 추천하는 방식</span>

```
Ru,i: 사용자 u, 아이템 i의 개인화된 예측 평점 값
Si,N: 아이템 i와 가장 유사도가 높은 TOP_N개 아이템의 유사도 벡터 
Ru,N: 사용자 u, 아이템 i와 가장 유사도가 높은 TOP_N개 아이템에 대한 실제 평점 벡터
   - N: 아이템의 최근접 이웃 범위 계수를 의미함
```

#### 사용자별 최적화된 평점 스코어 예측하는 사용자함수 생성

````python
ratings_matrix.head()
````

````python
item_sim_df.head(2)
````

````python
# rating_arr: 사용자-영화 평점 넘파이 행렬(ratings_matrix 데이터프레임을 넘파이 배열화)
# item_sim_arr: 영화 간 유사도를 가지는 넘파이 행렬(item_sim_df 데이터프레임을 넘파이 배열화)

def predict_rating(rating_arr, item_sim_arr):
    sum_S_i_N_R_u_N = rating_arr.dot(item_sim_arr)
    sum_S_i_N = np.array([np.abs(item_sim_arr).sum(axis = 1)])
    ratings_pred =  sum_S_i_N_R_u_N / sum_S_i_N
    
    return ratings_pred
````

````python
# ratings_matrix.values: ratings_matrix 전체 데이터프레임 넘파이배열화
# ratings_matrix.values: item_sim_df 전체 데이터프레임 넘파이배열화

ratings_pred = predict_rating(ratings_matrix.values, item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(ratings_pred, 
                                   index = ratings_matrix.index,
                                   columns = ratings_matrix.columns)
ratings_pred_matrix.head()
````

- 예측 평점이 사용자별 영화의 실제 평점과 영화의 코사인 유사도를 내적한 값이기 때문에 기존에 영화를 관람하지 않아 0에 해당했던 실제 영화 평점이 예측에서는 값이 부여되는 경우가 많이 발생함
- 따라서, 예측 평점이 실제 평점에 비해 작을 수 있음
- 이는 내적 결과를 코사인 유사도 벡터 합으로 나누었기 때문임

그렇다면, 예측 결과가 원래의 실제 평점과 얼마나 차이가 있는지 확인해보기

#### 예측 평점 및 실제 평점의 차이 확인하는 사용자 함수 생성

````python
from sklearn.metrics import mean_squared_error

def get_mse(pred, actual):
    # 평점이 있는 실제 영화만 추출
    ## nonzero(): ndarray 형태에서 0이 아닌 인덱스를 
    ## (array([  0,   0,   0, ..., 609, 609, 609], dtype=int64),
    ##  array ([  48,   66,  202, ..., 9712, 9715, 9716], dtype=int64)) 형태로 나타내줌
    
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    
    return mean_squared_error(pred, actual)


mse = get_mse(ratings_pred_matrix.values, ratings_matrix.values)

print('아이템 기반 모든 최근접 이웃 MSE: ', np.round(mse, 4))
````

- mse가 약 9.89임
- 실제 값과 예측값은 서로 스케일이 다르기 때문에 mse가 클 수 있고, 중요한 것은 mse를 감소시키는 방향으로 개선하는 것임

- predict_rating() 사용자 함수는 사용자별 영화의 예측 평점을 계산하기 위해 해당 영화와 다른 모든 영화 간의 유사도 벡터를 적용한 것임
- 많은 영화의 유사도 벡터를 이용하다 보니, 상대적으로 평점 예측이 떨어짐
- 따라서, 특정 영화와 가장 비슷한 유사도를 가지는 영화에 대해서만 유사도 벡터를 적용하는 함수로 변경하기

#### 특정영화와 유사도가 높은 영화에 대해 사용자별 최적화된 평점 스코어 예측하는 사용자함수 생성

````python
def predict_rating_topsim(ratings_arr, item_sim_arr, n = 20):
    
    # 사용자-아이템 평점 행렬 크기만큼 0으로 채운 예측 행렬 초기화
    pred = np.zeros(ratings_arr.shape)
    
    # 사용자-아이템 평점 행렬의 열 크기만큼 루프 수행
    for col in range(ratings_arr.shape[1]):
        
        # 유사도 행렬에서 유사도가 큰 순으로 n개 데이터 행렬의 인덱스 반환
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]        
        
        
        # 개인화된 예측 평점 계산 
        for row in range(ratings_arr.shape[0]):
            # 내적
            ## 아이템 유사도 데이터프레임 내에서 해당 영화와 top_n 영화들의 유사도 벡터
            ## 사용자-아이템 평점 데이터프레임 내에서 해당 사용자의 top_n 영화에 대한 평점 벡터  
            pred[row, col] = item_sim_arr[col,:][top_n_items].dot(ratings_arr[row, :][top_n_items].T)
            
            # 아이템 유사도 데이터프레임 내에서 해당 영화와 top_n 영화들의 유사도 벡터의 합
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col, :][top_n_items]))
            
    return pred
````

#### 예측 평점 및 실제 평점의 차이 확인하는 사용자 함수 생성

````python
ratings_pred = predict_rating_topsim(ratings_matrix.values, item_sim_df.values, n = 20)

mse = get_mse(ratings_pred, ratings_matrix.values)
print('아이템 기반 최근접 TOP20이웃 MSE: ', np.round(mse, 3))
````

- MSE가 3.69로 기존의 9.89보다 많이 향상됨
- 이제 특정 사용자에 대해 영화를 추천해 보겠음

````python
# 사용자9에 대해 영화 추천하기

user_rating_id = ratings_matrix.loc[9,:]
user_rating_id[user_rating_id > 0].sort_values(ascending = False)[:10]
````

- 반지의 제왕, 오스틴 파워 등 대작이나 어드벤처 영화, 코미디 영화를 전반적으로 흥행성이 좋은 영화에 높은 평점을 주고 있음
- 이 사용자에게 아이템 기반 협업 필터링을 통해 영화를 추천해보겠음

#### 사용자가 평점을 준 영화 제외한 영화를 리스트 객체로 반환하는 사용자함수 생성

````python
def get_unseen_movie(ratings_matrix, userId):
    # userId로 입력받은 사용자의 모든 영화 정보를 추출해 시리즈로 반환
    # 반환된 user_rating은 영화명(title)을 인덱스로 가지는 시리즈 객체임
    user_rating = ratings_matrix.loc[userId, :]
    
    # user_rating이 0보다 크면 기존에 관람한 영화이므로 대상 인덱스를 추출해 list 객체화
    already_seen = user_rating[user_rating > 0].index.tolist()
    
    # 영화명 list 객체화
    movie_list = ratings_matrix.columns.tolist()
    
    # already_seen에 해당하는 영화명을 지운 영화명을 list객체화
    movie_unseen_list = [movie for movie in movie_list if movie not in already_seen]
    
    return movie_unseen_list
````

#### 최종적으로 추천하는 사용자함수 생성
- 사용자가 평점을 준 영화를 제외
- 사용자별 아이템 유사도에 기반한 예측 평점 데이터 세트 이용

````python
def recommend_movie_by_userid(pred_df, userId, unseen_list, top_n = 10):
    # 예측평점이 데이터프레임에서 사용자 id 인덱스와 unseen_list로 들어온 영화명 컬럼을 추출
    # 가장 예측 평점이 높은 순으로 정렬함
    recommend_movies = pred_df.loc[userId, unseen_list].sort_values(ascending = False)[:top_n]
    
    return recommend_movies
````

````python
# 적용
## 사용자9에 대해 적용
ratings_pred = predict_rating_topsim(ratings_matrix.values, item_sim_df.values, n = 20)
pred_df = pd.DataFrame(ratings_pred, 
                       index = ratings_matrix.index,
                       columns = ratings_matrix.columns)
userId = 9
unseen_list = get_unseen_movie(ratings_matrix, userId)


recommend_movies = recommend_movie_by_userid(pred_df, userId, unseen_list, top_n = 10)
recommend_movies
````

# 잠재요인 협업 필터링
- 행렬분해 잠재 요인 협업 필터링은 SVD, NMF 등을 적용할 수 있는데, 일반적으로 행렬 분해에서는 SVD가 자주 사용됨
- 하지만, 사용자-아이템 평점 행렬에는 사용자가 평점을 매기지 않은 NULL 데이터가 많기 때문에, 주로 SGD(확률적 경사하강법), ALS 기반의 행렬 분해를 이용함
- <b>확률적 경사 하강법(SGD)</b>을 이용한 행렬 분해를 살펴보자면 <span style = 'background-color: #fff5b1'>P와 Q 행렬로 계산된 예측 R 행렬 값이 실제 R 행렬 값과 최소한의 오류를 가질 수 있도록 반복적으로 비용함수를 최적화함으로써 적합한 P와 Q 행렬을 유추하는 것이 알고리즘의 골자</span>

### SGD 기반 행렬 분해 사용자함수 생성

#### RMSE 구하는 사용자함수 생성

````python
def get_rmse(R, P, Q, non_zeros):
    error = 0
    
    # 두 개의 분해된 행렬 P와 Q.T의 내적으로 예측 R행렬 생성하기
    full_pred_matrix = np.dot(P, Q.T)
    
    # 실제 R행렬에서 NULL이 아닌 값의 위치 인덱스를 추출해서 실제 R행렬과 예측 행렬 간 RMSE 추출
    x_non_zero_ind = [i for i, j, r in non_zeros]
    y_non_zero_ind = [j for i, j, r in non_zeros]
    
    R_non_zeros = R[x_non_zero_ind, y_non_zero_ind]
    
    full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind]
    
    # 평균오차제곱 구하기
    mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
    rmse = np.sqrt(mse)
    
    return rmse
````

````python
def matrix_factorization(R, k, steps = 200, learning_rate = 0.01, r_lambda = 0.01):
    num_users, num_items = R.shape
    
    # P, Q 매트릭스의 크기를 지정하고 정규 분포를 가진 랜덤한 값으로 입력
    np.random.seed(1)
    
    P = np.random.normal(scale = 1./k, size = (num_users, k))
    Q = np.random.normal(scale = 1./k, size = (num_items, k))
    
    prev_rmse = 10000
    break_count = 0
    
    # R > 0인 행위치, 열위치, 값을 non_zeros 리스트 객체에 저장
    non_zeros = [(i, j, R[i,j]) for i in range(num_users) for j in range(num_items) if R[i, j] > 0]
    
    # SGD 기법으로 P와 Q 매트릭스를 계속 업데이트
    for step in range(steps):
        for i, j, r in non_zeros:
            
            # 실제 값과 예측 값의 차이인 오류값 구하기
            eij = r - np.dot(P[i,:], Q[j,:].T)
            
            # Regularization을 반영한 SGD 업데이트 공식 적용
            P[i, :] = P[i, :] + learning_rate * (eij * Q[j, :] - r_lambda * P[i, :])
            Q[j, :] = Q[j, :] + learning_rate * (eij * P[i, :] - r_lambda * Q[j, :])
            
        rmse = get_rmse(R, P, Q, non_zeros)
        
        if (step % 10) == 0:
            print(f'*****iteration step: {step}, rmse: {rmse}')
            
    return P, Q
````

### 데이터 불러오기

````python
import pandas as pd
import numpy as np

movies = pd.read_csv('dataset_ml_latest_small/movies.csv')
ratings = pd.read_csv('dataset_ml_latest_small/ratings.csv')

ratings_movies = pd.merge(movies, ratings, on = 'movieId')

display(movies)
display(ratings)
display(ratings_movies)
````

````python
# 사용자-아이템 행렬 만들기

ratings_matrix = pd.pivot_table(ratings_movies,
                                index = 'userId',
                                columns = 'title',
                                values = 'rating').fillna(0)

ratings_matrix
````

````python
P, Q = matrix_factorization(ratings_matrix.values, 
                            k = 50,
                            steps = 200,
                            learning_rate = 0.01,
                            r_lambda = 0.01)
pred_matrix = np.dot(P, Q.T)
pred_matrix
````

````python
# 예측 사용자-아이템 평점 행렬 생성 

ratings_pred_matrix = pd.DataFrame(data = pred_matrix,
                                   index = ratings_matrix.index,
                                   columns = ratings_matrix.columns)
ratings_pred_matrix.head(3)
````
