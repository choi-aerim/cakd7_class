```python
# 네이버 검색 API예제는 블로그를 비롯 전문자료까지 호출방법이 동일하므로 blog검색만 대표로 예제를 올렸습니다.
# 네이버 검색 Open API 예제 - 블로그 검색

import os
import sys
import urllib.request
import re
import pandas as pd

client_id = "vPYv0vamIlyLtJHkhBWS"
client_secret = "hkJBydGiAP"

query = urllib.parse.quote(input('>>검색어를 입력하세요.\n'))
url = "https://openapi.naver.com/v1/search/blog?query=" + query # json 결과

request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()

#빈 데이터프레임 작성
search_df = pd.DataFrame(columns = ['Title', 'Description', 'Link'])

if(rescode==200):
    response_body = response.read()
    response_dict = pd.read_json(response_body.decode('utf-8'))
#     print(response_dict)
    
    print(response_dict['items'])
    
    idx = 0
    items = response_dict['items']
    
    for items_index in range(len(items)):
        # 태그 모양 삭제
        p = re.compile('<b>|</b>')
        # 타이틀 뽑기
        title = re.sub(p,' ',items[items_index]['title'])
        # 링크 뽑기
        link = items[items_index]['link'] 
        # 본문 요약본 뽑기
        description = re.sub(p,' ',items[items_index]['description'])
        
        search_df.loc[idx] = [title, description, link]
        idx += 1
        

else:
    print("Error Code:" + rescode)
    
search_df
```

```python
import os
import sys
import urllib
import re
import pandas as pd

client_id = "vPYv0vamIlyLtJHkhBWS"
client_secret = "hkJBydGiAP"

query = urllib.parse.quote(input('>>검색어를 입력하세요\n'))
url = 'https://openapi.naver.com/v1/search/news?query=' + query

request = urllib.request.Request(url)
request.add_header('X-Naver-Client-Id', client_id)
request.add_header('X-Naver-Client-Secret', client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()    #수행이 잘되었는지 확인하는 코드를 받기

#빈 데이터프레임 생성
news_df = pd.DataFrame(columns = ['Press', 'NewsTitle', 'Description', 'Originalnews_link'])
idx = 0

if (rescode) == 200:
    response_body = response.read()
    response_dict = pd.read_json(response_body.decode('utf-8'))
    items = response_dict['items']
    print(items)
    
    for item_index in range(len(items)): 
        #뉴스 제목
        title = re.sub('&[\w]+;',' ',items.loc[item_index]['title'])
        
        #원문 제공 언론사(없으면 공백)
        try:
            press = re.search('\[(\w+)\]',title).group(1)
        except:
            press = ''
        #뉴스 제목에서 언론사 지우기
        title = re.sub('\[.*]','',title)
            
        #본문 내용
        description = re.sub('<.*>|&[\w]+;',' ',items.loc[item_index]['description'])
        
        #기사 원문 링크
        original_link = items.loc[item_index]['originallink']
        
        news_df.loc[idx] = [press, title, description, original_link]
        idx += 1

news_df
```

### 사용자함수로 만들기

```python
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus
import pandas as pd
import json


def baro_api(skey, res_id, encode):
    stype = 'json'
    url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
    parmas = f'?ServiceKey={skey}&type={stype}&resId={res_id}&'+ encode 
    query = url + parmas
    
    request = Request(query)
    request.get_method = lambda: 'GET'
    response_body = urlopen(request).read()
    
    response_dict = json.loads(response_body.decode('utf-8'))
    items = response_dict['body']['items']
    
    df = pd.DataFrame(items)
    return df
```
