### 평균구매주기기간 산정 사용자함수 정의

```python
from datetime import datetime

def out_q_df(df):
    df['TOTAL_CAT'] = df['BBCL'] + '_' + df['PRICE_TYPE']
    # 고객
    cust_li = df['CUSTNO'].unique().tolist()
    cust_li.sort()
    # 카테고리
    cate_li = df['TOTAL_CAT'].unique().tolist()
    cate_li.sort()
    
    # 최종 df 
    df_t = pd.DataFrame(index = cust_li, columns = cate_li)
    
    # 고객별, 카테고리별 df 반출
    for cate in cate_li:
        cate_df = df[df['TOTAL_CAT']==cate]
        
        # 고객리스트
        cust_li_ = cate_df['CUSTNO'].unique().tolist()
        cust_li_.sort()
        
        for cust in cust_li_:
            cust_df = cate_df[cate_df['CUSTNO']==cust]

            cat_date = cust_df['PURDATE'].unique().tolist()
            cat_date.sort(reverse=True)

            if len(cat_date) == 0:
                pass
            elif len(cat_date) == 1:
                mean_date = cat_date  #'just_one'
                df_t.loc[cust, cate] = mean_date
            else:
                date_li = []
                for day_ind in range(len(cat_date)-1):
                    day_after = cat_date[day_ind]
                    day_before = cat_date[day_ind + 1]

                    # 다음구매일 - 전구매일
                    date = datetime.strptime(str(day_after),'%Y%m%d') - datetime.strptime(str(day_before),'%Y%m%d')

                    date_li.append(date.days)
                # 평균 구매일    
                mean_date = sum(date_li) / len(date_li)

                df_t.loc[cust, cate] = mean_date
    return df_t
```

### 저장하기

```python
df_li = [q1_q3, q2_q4, q3_q5, q4_q6, q5_q7]
for i in range(5):
    df = out_q_df(df_li[i])
    df.to_csv(f'C:/cakd7/cakd7_class/github_구매감소예측프로젝트/second_projects/생성데이터/RFM/4. R_평균 구매주기 계산/R_total_Q{i+1}_Q{i+3}.csv')    
```

#### [yyyymmdd]를 전체기간으로 바꿔주는 사용자함수 생성

```python
import re

# 전체 데이터프레임 리스트
df_li = [q1_q3, q2_q4, q3_q5, q4_q6, q5_q7]

# 3분기간 전체 일수 리스트                                                                                     
days_li = [days_q123, days_q234, days_q345, days_q456, days_q567]

def quarter_period3(df_li, days_li, k): 
    
    # 사용자함수 정의
    ## 1. null값은 null값으로 두고
    ## 2. 1번 구매는 해당 기간으로 대체
    def replace_mean_purchase_period(x, i):
        if str(x).startswith('['):
            x = x.replace(x,str(days_li[i]))
            return x
        return x

    
    for i in range(k):
        df = df_li[i]

        # 날짜 단위 데이터 전체기간으로 변경 및 데이터타입을 float로 변경
        df = df.apply(lambda x: x.apply(lambda y: replace_mean_purchase_period(y,i))).astype(float)
        display(df)
        # 저장하기
#         df.to_csv(f'C:/cakd7/cakd7_class/github_구매감소예측프로젝트/second_projects/생성데이터/RFM/4. R_평균 구매주기 계산/R_total(한번구매보정)_Q{i+1}_Q{i+3}.csv')

quarter_period3(df_li, days_li, 5)
```

#### 등급화 후 저장

```python
# 전체 데이터프레임 리스트
df_li_2  = [q1_q3_2 , q2_q4_2 , q3_q5_2 , q4_q6_2 , q5_q7_2]
# 3분기간 전체 일수 리스트                                                                                     
days_li = [days_q123, days_q234, days_q345, days_q456, days_q567]


def grade_df(df_li_2, days_li):
    # 컬럼 이름 생성
    index_names = df_li_2[0].index
    column_names = df_li_2[0].columns

    # 등급을 채울 빈 데이터프레임 생성
    grade_df = pd.DataFrame(index = index_names, columns = column_names)


    # 등급화하는 함수 생성
    def grade_go(x, grade_기준, i):
        y = 0
        if x == days_li[i]: y = 6
        elif x <= grade_기준[0]: y = 1
        elif x <= grade_기준[1]: y = 2
        elif x <= grade_기준[2]: y = 3
        elif x <= grade_기준[3]: y = 4
        elif x >= grade_기준[3]: y = 5
        else: y = 7
        return y


    for i in range(len(df_li_2)):
        df = df_li_2[i]
        display(df)
        for col in range(len(df.columns)):
            # percentile을 기준으로 등급 나누기
            df_col = df.iloc[:,col]
            ## null값이랑 1번 구매한 사람 제외하고 등급 나누기
            df_col_nonull_nomax = df_col[(df_col != days_li[i]) & (df_col.isnull() == False)]

            grade_기준 = np.percentile(df_col_nonull_nomax, [10, 30, 70, 90])

            ## null값은 제일 마지막 등급, 1번 구매한 사람은 마지막에서 두번째 등급, 1 ~ 5 등급
            grade_df.iloc[:,col] = df.iloc[:,col].apply(lambda x: grade_go(x, grade_기준, i))

        display(grade_df)
        grade_df.to_csv(f'C:/cakd7/cakd7_class/github_구매감소예측프로젝트/second_projects/생성데이터/RFM/4. R_평균 구매주기 계산/R_total(한번구매보정)_Q{i+1}_Q{i+3}_등급화.csv')

grade_df(df_li_2, days_li)
```
