# 업종코드 리코딩

```python
import pandas as pd
```

```python
type = pd.read_excel('업종코드.xlsx', header = None)
type
```

### 대분류/중분류 업종코드 데이터프레임화

#### 업종코드 데이터프레임화

```python
# 업종코드별 세부타입 데이터프레임 만들기
type_clean = pd.DataFrame(columns = ['code','type'])

code_list = []
type_list = []

for i in type[0]:
    code_str = i.split(' ')[-2]## 업종코드 리코딩### 업종코드 데이터프레임화
    type_str = i.split(' ')[-1]
    
    code_list.append(code_str)
    type_list.append(type_str)
    
    
type_clean['code'] = code_list
type_clean['type'] = type_list

type_clean
```

#### 중분류 업종코드 데이터프레임화

```python
# 세부업종 코드 데이터프레임

detail_type_clean = type_clean[7:]
detail_type_clean.rename(columns = {'type': 'detail_type'}, inplace = True)
detail_type_clean#### 업종코드 데이터프레임화
```

```python
# 코드를 문자 ->숫자화

detail_type_clean['code'] = detail_type_clean['code'].astype(int)
```

```python
detail_type_clean['code'].dtype
```

#### 대분류 업종코드 데이터프레임화

```python
#업종별 코드

type_clean = type_clean[:7]
type_clean
```

```python
# 코드를 문자 ->숫자화

type_clean['code'] = detail_type_clean['code'].astype(int)
```
