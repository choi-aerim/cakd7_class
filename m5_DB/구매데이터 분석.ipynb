### 구매연도, 월, 일 컬럼 생성

```python
purchase['구매연도'] = purchase['구매일자'].apply(lambda x: int(str(x)[:4]))
purchase['구매월'] = purchase['구매일자'].apply(lambda x: int(str(x)[4:6]))
purchase['구매일'] = purchase['구매일자'].apply(lambda x: int(str(x)[6:]))

def q(x):
    y = 0
    if x in (1,2,3): y = 1
    elif x in (4,5,6): y = 2
    elif x in (7,8,9): y = 3
    elif x in (10,11,12): y = 4
    return y
purchase['구매 분기'] = purchase['구매월'].apply(lambda x: q(x))
purchase
```

```python
def season(x):
    y = 0
    if x in (12,1,2): y = '겨울'
    elif x in (3,4,5): y = '봄'
    elif x in (6,7,8): y = '여름'
    elif x in (9,10,11): y = '가을'
    return y
purchase['구매 계절'] = purchase['구매월'].apply(lambda x: season(x))
purchase.head()
```

### 연도별 구매금액 변화

```python
purchase.groupby('구매연도').sum()['구매금액']
```

```python
pur2014 = purchase.groupby('구매연도').sum()['구매금액'][2014]
pur2015 = purchase.groupby('구매연도').sum()['구매금액'][2015]

pur_rate = (pur2015 - pur2014)/pur2014
round(pur_rate,4)
```

5.4% 증가

### 연도/분기별 구매금액 변화

```python
quarter_rate = pd.pivot_table(data = purchase,
                              index = '구매연도',
                              columns = '구매 분기',
                              values = '구매금액',
                              aggfunc = 'sum')
quarter_rate
```

```python
# 증감율
round(quarter_rate.pct_change(),3)*100
```

### 연도/계절별 구매금액 변화

```python
season_rate = pd.pivot_table(data = purchase,
                             index = '구매연도',
                             columns = '구매 계절',
                             values = '구매금액',
                             aggfunc = 'sum')
season_rate
```
