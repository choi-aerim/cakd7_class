## 함수 생성

```python
def pivot(df, aggfunc):
    # 분기 변경
    df['Q_year'] = df['PURYEAR'].astype(str)+ '_' + df['QUARTER'].astype(str)
    df['Q_year']

    # 년도
    pivot_df_year = pd.pivot_table(df,
                             values = 'PURAMT',
                             index = 'CUSTNO',
                             columns = ['PURYEAR', 'BBCL', 'PRICE_TYPE'],
                             aggfunc = aggfunc)
    pivot_df_year.fillna(0, inplace = True)
    columns_name = []
    for indexs in pivot_df_year.columns:
         columns_name.append('_'.join([aggfunc] + [str(idx) for idx in indexs]))
    pivot_df_year.columns = columns_name
    
    # 년도 총액
    pivot_df_label_year = pd.pivot_table(df,
                                        values = 'PURAMT',
                                        index = 'CUSTNO',
                                        columns = ['PURYEAR'],
                                        aggfunc = 'sum')
    pivot_df_label_year.fillna(0, inplace = True)
    
    # 반기
    pivot_df_half = pd.pivot_table(df,
                                     values = 'PURAMT',
                                     index = 'CUSTNO',
                                     columns = ['HALF_YEAR', 'BBCL', 'PRICE_TYPE'],
                                     aggfunc = aggfunc) # 대분류하는것
    pivot_df_half.fillna(0, inplace = True)
    # 컬럼 변경
    columns_name = []
    for indexs in pivot_df_half.columns:
         columns_name.append('_'.join([aggfunc, '반기'] + [str(idx) for idx in indexs]))
    pivot_df_half.columns = columns_name
    
    # 반기 총액
    pivot_df_label_half = pd.pivot_table(df,
                                        values = 'PURAMT',
                                        index = 'CUSTNO',
                                        columns = ['HALF_YEAR'],
                                        aggfunc = 'sum')
    pivot_df_label_half.fillna(0, inplace = True)
    
    # 분기
    pivot_df_q_ = pd.pivot_table(df,
                               values = 'PURAMT',
                               index = 'CUSTNO',
                               columns = ['Q_year', 'BBCL', 'PRICE_TYPE'],
                               aggfunc = aggfunc)
    pivot_df_q_.fillna(0, inplace = True)
    # 컬럼 변경
    columns_name = []
    for indexs in pivot_df_q_.columns:
         columns_name.append('_'.join([aggfunc, '분기'] + [str(idx) for idx in indexs]))
    pivot_df_q_.columns = columns_name
    
    # 분기 총액
    pivot_df_label_q = pd.pivot_table(df,
                                values = 'PURAMT',
                                index = 'CUSTNO',
                                columns = ['Q_year'],
                                aggfunc = 'sum')
    pivot_df_label_q.fillna(0, inplace = True)
    
    pivot_total_ = pd.concat([pivot_df_year, pivot_df_half, pivot_df_q_], axis = 1)
    
    def minus(x):
        if x > 0: y = 1
        else: y = 0
        return y
  
    # 년도기준 증가
    pivot_total_['year_up'] = pivot_df_label_year[2015] -  pivot_df_label_year[2014]
    pivot_total_['year_label'] = pivot_total_['year_up'].apply(lambda x : minus(x))
    # 반기기준 증가
    pivot_total_['half_up'] = pivot_df_label_half[4] -  pivot_df_label_half[1]
    pivot_total_['half_label'] = pivot_total_['half_up'].apply(lambda x : minus(x))
    # 분기기준 증가
    pivot_total_['q_up'] = pivot_df_label_q['2015_4'] -  pivot_df_label_q['2014_1']
    pivot_total_['q_label'] = pivot_total_['q_up'].apply(lambda x : minus(x))
    
    total_df = pivot_total_.copy()
    return total_df
```

## 함수 대입

```python
purchase_pviot_cnt = pivot(purchase_real, 'count')
purchase_pviot_cnt
```
