### 고/중/저가별 매출액 변화

```python
for PRICE_TYPE in plus_df['PRICE_TYPE'].unique():
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot(111)
    pivot = pd.pivot_table(plus_df[plus_df['PRICE_TYPE'] == PRICE_TYPE], 
                   index = 'PRICE_TYPE', 
                   columns = ['PURYEAR','PURMONTH'],
                   values = 'PURAMT', aggfunc = 'sum')
    ax.plot(pivot[2014].transpose(), marker = 'o', label = '2014년')
    ax.plot(pivot[2015].transpose(), marker = 'o', label = '2015년')
    ax.set_xticks(list(range(1,13)))
    ax.set_xlabel(PRICE_TYPE)
    ax.legend()
```

```python
# 대분류 코드별
BCLS = plus_df['BCL'].unique()
for BCL in BCLS:
    
    pivot = pd.pivot_table(plus_df[(plus_df['BCL'] == BCL)],
                            index = ['BCL','PRICE_TYPE'],
                            columns = ['PURYEAR','PURMONTH'],
                            values = 'PURAMT',
                            aggfunc = 'sum')
    fig = plt.figure(figsize = (15,3))
    try:
        ax1 = fig.add_subplot(131)
        ax1.plot(pivot.loc[(BCL, '고가')][2014].transpose(), marker = 'o', color = 'red', label = '2014_고가')
        ax1.plot(pivot.loc[(BCL, '고가')][2015].transpose(), marker = 'o', color = 'orange', label = '2015_고가')
        ax1.set_xticks(list(range(1,13)))
        ax1.set_xlabel(f'{BCL}')
        ax1.legend()
    except:
        pass
    try:
        ax2 = fig.add_subplot(132)
        ax2.plot(pivot.loc[(BCL, '중가')][2014].transpose(), marker = 'o', color = 'red', label = '2014_중가')
        ax2.plot(pivot.loc[(BCL, '중가')][2015].transpose(), marker = 'o', color = 'orange', label = '2015_중가')
        ax2.set_xticks(list(range(1,13)))
        ax2.set_xlabel(f'{BCL}')
        ax2.legend()
    except:
        pass
    try:
        ax3 = fig.add_subplot(133)
        ax3.plot(pivot.loc[(BCL, '저가')][2014].transpose(), marker = 'o', color = 'red', label = '2014_저가')
        ax3.plot(pivot.loc[(BCL, '저가')][2015].transpose(), marker = 'o', color = 'orange', label = '2014_저가')
        ax3.set_xticks(list(range(1,13)))
        ax3.set_xlabel(f'{BCL}')
        ax3.legend()
    except:
        pass    
```
