#### 아로마 키값 확인

```python
key_all = []

for idx in range(len(wine_df['아로마 딕셔너리'])):
    for key in eval(wine_df.loc[idx,'아로마 딕셔너리']).keys():
        key_all.append(key)
    
key_all = list(set(key_all))
```
