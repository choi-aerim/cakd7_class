### 데이터 전처리: 신세계 픽 버리기

```python
shin_wine['나라, 와이너리'].unique()
```

```python
import numpy as np
```

```python
delete_idx  = shin_wine[shin_wine['나라, 와이너리'].isin(['칠레 > 센트럴 밸리 / 신세계엘앤비 픽','이태리 > 피에몬테 / 신세계엘앤비 픽', '스페인 > 까바 / 신세계엘앤비 픽',
       '프랑스 > 샴페인 / 신세계엘앤비 픽', '미국 > 캘리포니아 / 신세계엘앤비 픽', '미국 > 나파밸리 / 신세계엘앤비 픽', '스페인 > 갈리시아 / 신세계엘앤비 픽', 
       '포르투갈 > 리스보아 / 신세계엘앤비 픽', '프랑스 > 다수지역 / 신세계엘앤비 픽','스페인 > 다수지역 / 신세계엘앤비 픽',
       '독일 > 다수 지역 / 신세계엘앤비 픽','칠레 > 다수지역 / 신세계엘앤비 픽', '프랑스 > 보르도 / 신세계엘앤비 픽',
       '프랑스 > 프로방스 / 신세계엘앤비 픽','프랑스 > 부르고뉴 / 신세계엘앤비 픽','프랑스 > 남프랑스 / 신세계엘앤비 픽'])].index

shin_wine = shin_wine.drop(index = delete_idx) 
shin_wine = shin_wine.reset_index().drop('index', axis = 1)
shin_wine.to_pickle('신세계wine_delete.pkl')
shin_wine.to_csv('신세계wine_delete.csv')
```
