```python
#컬럼 개수가 45개 이상이고 서울을 포함한 행  내용  뽑아내기 (나중에 직접 추가하게,,)

import re

with open('원천정산경락가격원시데이터_202206.csv', encoding='utf-16' ) as f:
    lines = f.readlines()
    i = 0
    j = 0
    for line in lines:
        i += 1
        if len(line.split(','))>45 and len(re.findall( '서울',line))>0:
            length=len(line.split(','))
            print(f'{i}번째 행: {length}', line)
            j+=1
print(j)   
```

# 서울시 생필품 농수축산물 가격 정보

```python
#서울시 생필품 농수축산물 가격정보: raw 데이터

import pandas as pd
price_retail = pd.read_csv('서울시 생필품 농수축산물 가격.csv',encoding = 'utf-8')
price_retail
```

```python
price_retail['품목 이름'].unique()
```

```python
#서울시 생필품 농수축산물 가격정보: 판매규격 뽑아내기(냉동, 생물, 국산, 수입, n개)

import re

for i in range(len(price_retail)):
    try:
        price_retail.loc[i,'판매규격'] = re.search('\((.*)\)',str(price_retail.loc[i,'품목 이름'])).group(1)
    except:
        price_retail.loc[i,'판매규격'] = ''
    price_retail.loc[i,'품목 이름'] = re.sub('\(.*\)','',str(price_retail.loc[i,'품목 이름']))

price_retail
```
