```python
import pandas as pd

landmark = pd.read_csv('서울시 관광 명소.csv')
landmark
```

```python
landmark = landmark[landmark['언어'] == 'ko']
landmark.reset_index(inplace = True)
landmark
```

```python
landmark.drop('index', axis = 1, inplace = True)
landmark
```

```python
landmark_name = list(landmark['상호명'])
landmark_name
```

```python
import re

landmark_name2 = []
for land in landmark_name:
    land = re.sub('\(.*\)','',str(land))
    landmark_name2.append(land)
    
landmark_name2
```
