### 데이터 전처리

```python
import pandas as pd
import numpy as np
import random
```

```python
#고객별 속성 데이터프레임 만들기

cusno=np.arange(1000,2000)
#cusno=random.sample(list(range(1000,9000)),1000)
gender=np.random.choice([0,1],size=1000)
age=np.random.randint(10,81,size=1000)
region=np.random.randint(1,11,size=1000)
```

```python
cusno=pd.Series(cusno)
gender=pd.Series(gender)
age=pd.Series(age)
region=pd.Series(region)

#시리즈 합쳐서 데이터프레임 만들기
customer=pd.concat([cusno,gender,age,region],axis=1)

#컬럼명 지정
customer.columns=['cusno','gender','age','region']

customer.head()
```

```python
#거래 데이터프레임 만들기

product_20=np.random.choice([10,20,30],size=1000)
product_21=[random.randrange(10,40,10) for i in range(1000)]
price_20=np.random.uniform(1000,50000,size=1000).round(2)
price_21=np.random.uniform(1000,50000,size=1000).round(2)
quantity_20=np.random.randint(1,100,size=1000)
quantity_21=np.random.randint(1,100,size=1000)
amount_20=price_20*quantity_20
amount_21=price_21*quantity_21
sales=np.where(price_21>price_20,1,0)

#빈 데이터프레임 만들기
purchase=pd.DataFrame(columns=['product_20','product_21','price_20','price_21',
                              'quantity_20','quantity_21','amount_20','amount_21',
                              'sales'
])

#데이터 채우기
purchase['product_20']=product_20
purchase['product_21']=product_21
purchase['price_20']=price_20
purchase['price_21']=price_21
purchase['quantity_20']=quantity_20
purchase['quantity_21']=quantity_21
purchase['amount_20']=amount_20
purchase['amount_21']=amount_21
purchase['sales']=sales

purchase.head()
```

```python
#데이터프레임 통합: 공통 변수가 없기 때문에 인덱스 행 기준으로 결합하는 join함수의 outer을 활용함

df=customer.join(purchase, how='outer')
df.head()
```

```python
#제품의 세부코드(1~3) 변수 추가
#고가제품 10(1:, 2:, 3:)
#중가제품 20(1:, 2:, 3:)
#저가제품 30(1:, 2:, 3:)

df['detail_20']=np.random.randint(1,4,size=1000)
df.head()
```

```python
#21년도와 20년도의 구매액 차이 변수 추가

df['amount_df']=df.amount_21-df.amount_20
df.head()
```

```python
#gender value 문자화

df['gender'].replace({0:'male',1:'female'}, inplace=True)
df.head()
```

```python
#region value 문자화

df['region'].replace({1:'서울', 
                      2:'경기',
                      3:'부산',
                      4:'강원',
                      5:'충북',
                      6:'충남',
                      7:'전남',
                      8:'전북',
                      9:'경남',
                      10:'경북',                      
                     },inplace=True)
df.head()
```

```python
#age 범주화

def age_category(x):
    y=0
    if x<20:
        y=10
    elif x<30:
        y=20
    elif x<40:
        y=30
    elif x<50:
        y=40
    elif x<60:
        y=50
    else:
        y=60
    return y

df['age_c']=df['age'].apply(age_category)
```
