### 중복된 행정동(읍면동명), 법정동(동리명)이 있는지 확인

```python
#행정구 개수 확인

len(set(df_need.set_index('시군구명').index))
```

```python
#중복되는 행정동명 추출

dong = []
gudong_dict = {}
for i in set(df_need.set_index('시군구명').index):
    dong_raw = list(set(df_need.loc[df_need['시군구명'] == i, '읍면동명']))
    for j in dong_raw:
        dong.append(j)
    gudong_dict[i] = dong_raw

    s_dong = set(dong)
    raw_dong = dong
    if len(s_dong) != len(raw_dong):
        for k in s_dong:
            raw_dong.remove(k)
        print(raw_dong)
            
```

- 신사동은 강남구, 관악구에 각각 존재함

```python
#중복되는 법정동명 추출

dong = []
gudong_dict = {}
for i in set(df_need.set_index('시군구명').index):
    dong_raw = list(set(df_need.loc[df_need['시군구명'] == i, '동리명']))
    for j in dong_raw:
        dong.append(j)
    gudong_dict[i] = dong_raw

    s_dong = set(dong)
    raw_dong = dong
    if len(s_dong) != len(raw_dong):
        for k in s_dong:
            raw_dong.remove(k)
        print(raw_dong)
            
```

- 신정동은 마포구, 양천구에 존재
- 신사동은 강남구, 은평구에 존재

### 데이터 전처리

#### 행정동코드, 법정동코드 단위 맞추기

```python
import numpy as np
df2_hund = df_need.copy()
df2_hund['행정동코드'] = (df2_hund['행정동코드']/100).astype(np.int64)
df2_hund['법정동코드'] = (df2_hund['법정동코드']/100).astype(np.int64)
df2_hund
```

#### 행정동 별 법정동 개수 매핑

```python
df2_count = pd.merge(df2_hund, df2_hund[['행정동코드','법정동코드']].groupby('행정동코드').count().rename(columns = {'법정동코드': '법정동개수'}).reset_index(), how = 'left')
df2_count
```
